```r
library("stylo")
```

```r
articles <- load.corpus(files = "all", corpus.dir = "NINA/", encoding = "UTF-8")
```

```r
tokenized.corpus <- txt.to.words(articles, preserve.case = FALSE)
grams <- txt.to.features(tokenized.corpus, ngram.size = 3, features = "c")
```

```r
data <- grams
clevel <- 25
sliced.corpus <- make.samples(data, sampling = "random.sampling", sample.size = 225)
```

```r
help(stylo)
```

```r
frequent.features <- make.frequency.list(data)
freqs <- make.table.of.frequencies(data, features = frequent.features)
culled.freqs <- perform.culling(freqs, culling.level = clevel)
my.stylometric.test = stylo(frequencies = culled.freqs, gui = TRUE, path = NULL)
```

```r
summary(my.stylometric.test)
my.stylometric.test$distance.table
```
