```python
import pandas as pd
import networkx as nx
import dhlab.graph_networkx_louvain as gnl
```

```python
graph_data = pd.read_csv("Kent_CA_100_MFWs_Culled_0__Classic Delta_EDGES.csv")
```

```python
adj = graph_data.pivot("Source", "Target", "Weight").fillna(0)

G = nx.from_pandas_adjacency(adj)

gnl.show_graph(G, fontsize=10)
```

```python
gnl.show_communities(G)
```
